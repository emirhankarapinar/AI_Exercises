```python
def degisken(vari):
    cat=titanic_train[vari]
    sayi=cat.value_counts()
    plt.figure(figsize=(9,3))
    plt.bar(sayi.index,sayi)
    plt.xticks(sayi.index,sayi.index.values)
    plt.ylabel("frekans")
    plt.title(vari)
    plt.show()
    print(vari,sayi)
```

```python
category1=["Survived","Sex","Pclass","Embarked","SibSp","Parch"]
```

```python
for i in category1:
   degisken(i)
```

```python
def hist(vari):
    plt.figure(figsize=(9,3))
    plt.hist(titanic_train[vari])
    plt.xlabel(vari)
    plt.ylabel("frekans")
    plt.title(vari)
    plt.show()
    
```

```python
sayisal=["Fare","Age","PassengerId"]
for i in sayisal:
      hist(i)
```
